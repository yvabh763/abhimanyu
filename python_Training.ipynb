```python
#tuple has less time and space complexity than list
import time
import sys
L=list(range(100000))
T=tuple(range(100000))
start= time.time()
for i in L:
  i*10
print("List time= ",time.time()-start)
start= time.time()
for i in T:
  i*10
print('Tuple time= ',time.time()-start)
print('List size',sys.getsizeof(L))
print('Tuple size',sys.getsizeof(T))
```
