```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\user\OneDrive\Desktop\python folder\dataset\Student_data.csv')
df.head()
```

```python
internet_counts = df['internet'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(internet_counts, labels=internet_counts.index, autopct='%1.1f%%', startangle=90, colors=['#99ff99','#ffcc99'])
plt.title('Internet Access Among Students')
plt.show()
```

```python
avg_grade = df.groupby('studytime')['G3'].mean()

plt.figure(figsize=(8,5))
plt.bar(avg_grade.index.astype(str), avg_grade.values, color='orange')

plt.title('Average Final Grade by Study Time')
plt.xlabel('Study Time (1-4)')
plt.ylabel('Average Final Grade (G3)')
plt.show()
```
