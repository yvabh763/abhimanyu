```python
import time
import sys

del list
L = list(range(100000))   # List of 100,000 integers
T = tuple(range(100000))  # Tuple of 100,000 integers

# Measure time to loop through list
start = time.time()
for i in L:
    i * 10
print("List time= ", time.time() - start)

# Measure time to loop through tuple
start = time.time()
for i in T:
    i * 10
print("Tuple time= ", time.time() - start)

# Memory comparison
print("List size =", sys.getsizeof(L))
print("Tuple size =", sys.getsizeof(T))
```
